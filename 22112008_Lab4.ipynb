Let us see which method of scaling should we use?

1. MinMaxScaler - MinMaxScaler can be used here as we know that the range of values lie between 0 - 100. Also, we do not need the outliers in this problem statement
2. Max Absolute Normalisation - We saw that the number of 0 values in the dataset is not more. So we need not use this method
3. Robust Scaling - The outliers are not important in this scenario, hence we will not use Robust Scaling.
4. StandardScaler - If we want the outliers to be there we can use this scaling.

As outliers are not important with respect to my problem statement, I will be using MinMaxScaler.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = knn.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
scores = []
for i in range(1,100):
    kn = KNeighborsClassifier(n_neighbors = i)
    kn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,100),scores)
```
