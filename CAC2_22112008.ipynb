# About the dataset

Sure, here's an explanation of each column in the QS Ranking dataset:

1. **sustainability**: This column represents the sustainability score of the university. Sustainability here typically refers to environmental sustainability efforts and practices.

2. **employment_outcomes**: This column indicates the employment outcomes score of the university, reflecting how well graduates from the university fare in terms of finding employment after graduation.

3. **international_research_network**: This column represents the score for the international research network of the university, which reflects the university's collaborations and partnerships with other institutions worldwide.

4. **international_students_ratio**: This column denotes the ratio of international students at the university, indicating the extent to which the university attracts students from different countries.

5. **international_faculty_ratio**: This column indicates the ratio of international faculty members at the university, showing the diversity of the faculty in terms of nationalities.

6. **citations_per_faculty**: This column represents the number of citations per faculty member at the university, indicating the impact and quality of the research conducted at the university.

7. **faculty_student_ratio**: This column denotes the ratio of faculty members to students at the university, indicating the level of personalized attention and support available to students.

8. **employer_reputation**: This column represents the employer reputation score of the university, reflecting how well the university is regarded by employers when hiring graduates.

9. **academic_reputation**: This column indicates the academic reputation score of the university, reflecting the perception of academic quality and excellence by academics and experts in the field.

10. **overall**: This column represents the overall score of the university in the QS Rankings, which is a composite score based on various factors including those mentioned above.

11. **QS_24**: This column represents the position of the university in the QS Rankings.

12. **location**: This column provides the location of the university.

13. **name**: This column provides the name of the university.

# Predicting Sustainability

For predicting sustainability to encode location we will use target encoding as target encoding encodes by taking the mean of target variable for that group and here the higher the target variable the better it is.

```python
target_encoding = df1.groupby('location')['sustainability'].mean()
df1['location'] = df1['location'].map(target_encoding)
```

```python
df1.head()
```

### Scaling

MinMaxScaler - It cannot be used as the minimum and maximum value are not sure for the above features.
MaxAbsolute - It cannot be used here as the data is not sparse
Robust Scaling - It cannot be used here as the importance in the problem statement is not on the outliers.
Hence StandardScaler will be used

```python
X,y = df1.drop(columns=['sustainability']),df1['sustainability']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df1)
df1 = pd.DataFrame(scaled_data, columns=df1.columns)
df1.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state=0)
```

### KNN Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
```

```python
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
y_pred = knn.predict(X_test)

print('MAE',mean_absolute_error(y_test,y_pred))
print('MSE',mean_squared_error(y_test,y_pred))
print('R2 Score',r2_score(y_test,y_pred))
```

The best suited metrics is R2 Score as it tells the variation in the features.

```python
from sklearn.model_selection import cross_val_score
cross_val_score(knn, X_train, y_train, cv=5, scoring='r2')
```
