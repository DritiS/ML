```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException

# Configuration
base_url = "https://www.topuniversities.com/world-university-rankings"

# Initialize WebDriver
options = webdriver.ChromeOptions()
# options.headless = True  # Uncomment if you do not need a browser UI.
driver = webdriver.Chrome(options=options)
driver.maximize_window()

# Initialize an empty list to store the data
data_list = []

def get_university_data(driver, url):
    try:
        driver.get(url)
        WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".circle .score")))

        data = {}
        metrics = [
            "sustainability", "employment_outcomes", "international_research_network",
            "international_students_ratio", "international_faculty_ratio", "citations_per_faculty",
            "faculty_student_ratio", "employer_reputation", "academic_reputation", "overall"
        ]

        # Extract data from the circles on the page
        circles = driver.find_elements(By.CSS_SELECTOR, ".circle .score")
        for i, metric in enumerate(metrics):
            data[metric] = circles[i].text if i < len(circles) else None

        # Additional details
        data["QS_24"] = driver.find_element(By.CSS_SELECTOR, "#wur-tab .latest_rank").text
        data["location"] = driver.find_element(By.CSS_SELECTOR, "#p2-campus-location .active").text
        data["name"] = driver.find_element(By.CSS_SELECTOR, "#p2-overview .univ-block-heading").text
        
        return data

    except (NoSuchElementException, TimeoutException) as e:
        print(f"Error fetching data for URL {url}: {e}")
        return None  # Return None or {}

# Scrape multiple pages
try:
    for page in range(0, 15):
        driver.get(f"{base_url}?page={page}")
        WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".uni-link")))
        links = driver.find_elements(By.CSS_SELECTOR, 'a.uni-link')
        urls = [link.get_attribute('href') for link in links]

        for url in urls:
            university_data = get_university_data(driver, url)
            if university_data:
                data_list.append(university_data)  # Append the data to the list
finally:
    driver.quit()

# Convert the list of dictionaries to a Pandas DataFrame
df = pd.DataFrame(data_list)

# Display the DataFrame
print(df)
```
