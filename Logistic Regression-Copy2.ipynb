```python
import pandas as pd
df = pd.read_csv('Student Engagement Level-Binary.csv')
df
```

```python
df = df.drop('Student ID',axis=1)
```

```python
df
```

```python
df.isna().mean()*100
```

```python
# We use label encoding as we need to encode target variable
 
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 


df['Engagement Level']= label_encoder.fit_transform(df['Engagement Level']) 

df['Engagement Level'].unique() 
```

```python
df
```

```python
for column_name in df.columns:
    column = df[column_name]
    # Get the count of Zeros in column 
    count = (column == 0).sum()
    print(count)
```

```python
# There are many columns which have a lot of zero values. So we will use Max Absolute Normalization.
columns = ['# Forum Reads','# Forum Posts','# Quiz Reviews before submission','Assignment 1 lateness indicator','Assignment 2 lateness indicator','Assignment 3 lateness indicator']
for i in columns:
    df[i] = df[i] / df[i].abs().max()
df
```
